### 데이터 형 변환 출력 포맷팅

```python
# 문자열로 입력받은 값들을 정수형으로 변환
a = int(input('첫 번째 정수 입력: '))
b = int(input('두 번째 정수 입력: '))

print('{0} * {1} = {2:,}'.format(a, b, a*b))
```

```python
# 문자열로 입력받은 값들을 실수형으로 변환
c = float(input('첫 번째 실수 입력: '))
d = float(input('두 번째 실수 입력: '))

print('{0} * {1} = {2:,}'.format(c, d, c*d))
```

```python
# 문자열로 입력받은 값들을 실수형으로 변환
# f-string formatting 사용
c = float(input('첫 번째 실수 입력: '))
d = float(input('두 번째 실수 입력: '))
g = c * d

# g 변수에 truncate 된 형태로 저장되어 있는 것이 아니라, 화면에 표시할 때만 자리수를 지정해서 출력.
print(f'{c:.2f} * {d:.2f} = {g:,.2f}')
```

### 원의 둘레와 넓이 계산

```python
# 반지름 실수값을 입력 받아, 원의 둘레와 넓이 계산
# 소숫점 이하 1자리로 표시. 원의 넓이는 천 단위 구분기호(,)를 표시
# f-string formatting 사용

r = float(input('반지름 입력(cm): '))

print(f"\n반지름: {r:.1f}cm")
print(f"원의 둘레: {r * 2 * 3.14:.1f}cm")
print(f"원의 넓이: {r * r * 3.14:,.1f}cm²")
```

### 중첩 if 문
- 출생 연도 끝자리와 나이를 입력받아 마스크 구매 가능 요일 출력하기

```python
year = int(input("출생 연도 끝자리를 입력하세요: "))
age = int(input("현재 나이를 입력하세요: "))

# 65세 이상인 분들을 가장 먼저 처리
if age >= 65:
    print('언제든지 구매 가능합니다.')
# 65세 미만인 분들은 출생 연도 끝자리 수에 따라 구매 가능한 요일을 출력
elif year in [1, 6]:
    print('구매 가능한 요일: 월요일')
elif year in [2, 7]:
    print('구매 가능한 요일: 화요일')
elif year in [3, 8]:
    print('구매 가능한 요일: 수요일')
elif year in [4, 9]:
    print('구매 가능한 요일: 목요일')
else:
    print('구매 가능한 요일: 금요일')
```

- 차량 2부제: Day와 차량번호를 입력받아 운행 여부 출력하기

```python
today = int(input('오늘 날짜 입력: '))

if 0 <= today <= 31:
    rego_str = input('차량번호 입력: ')
    if len(rego_str) != 4:
        print('차량번호 오류')
    else:
        rego = int(rego_str)
        if rego % 2 == 0: # 짝수
            print('짝수 차량 운행날')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 가능')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 불가능')
        else: # 홀수
            print('홀수 차량 운행날')
            if today % 2 == 0:
                # 오늘 날짜가 짝수이면
                print('운행 불가능')
            else:
                # 오늘 날짜가 홀수이면
                print('운행 가능')    
else:
    print('날짜 입력 오류')
```

#### if elif else
- 키와 몸무게를 입력받아 BMI 지수 및 비만 유무 출력하기

```python
height = float(input('키 입력(cm): '))
weight = int(input('몸무게 입력(kg): '))

bmi = weight / ((height ** 2) / 10000)
if bmi >= 23:
    print(f'BMI 지수는 {bmi:.2f}로 "과체중"입니다.')
elif bmi >= 18.5:
    print(f'BMI 지수는 {bmi:.2f}로 "정상체중"입니다.')
else:
    print(f'BMI 지수는 {bmi:.2f}로 "저체중"입니다.')
```

### while 문
- 사용자로부터 정수값을 입력받아 1부터 입력받은 숫자까지의 합 구하기

```python
endValue = int(input("합을 원하는 정수값 입력: "))

i = 1
total = 0
while i <= endValue:
    total += i
    i += 1
print(f"1부터 {endValue}까지의 합은 {total:,}입니다.")
```
