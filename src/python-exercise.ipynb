### MISC
- upper / lower / swapcase

```python
print('Hello World!'.upper())
print('Hello World!'.lower())
print('Hello World!'.swapcase())
```

### strip

```python
print(len(' Hello World!  '))
print(len(' Hello World!  '.lstrip()))
print(len(' Hello World!  '.rstrip()))
print(len(' Hello World!  '.strip()))
```

### split

```python
print('Hello World!'.split())
print('Hello  World!'.split())
print(' Hello World!'.split())
print('Hello World! '.split())
```

### JSON

```python
import json

# 파이썬 딕셔너리 데이터
data = {
    'name': '홍길동',
    'age': 30,
    'city': 'Seoul'
}

# 딕셔너리를 JSON 문자열로 변환
print(json.dumps(data, ensure_ascii=True))
print(json.dumps(data, ensure_ascii=False)) # ensure_ascii=False 옵션을 사용하여 한글이 깨져 보이지 않도록 함

json_string = json.dumps(data, ensure_ascii=False)
print(f'JSON 문자열 : {json_string}')
print(type(json_string))

# JSON 문자열을 다시 파이썬 딕셔너리 데이터 타입으로 변환
parsed_data = json.loads(json_string)
print(f'파이썬 데이터 타입 : {type(parsed_data)}')
print(f'파이썬 데이터 : {parsed_data}')
```

### JSON 파일 입출력

```python
# 파일 쓰기
with open('../data/example1.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
# 파일 읽기
with open('../data/example1.json', 'r', encoding='utf-8') as f:
    content = json.load(f)
    print(content)
    print(type(content))
    print('나이:', content['age'])
```

### csv

```python
import csv

csv_data = [
    ['name', 'age', 'city'],
    ['홍길동', 30, 'Seoul'],
    ['김철수', 25, 'LA'],
    ['이영희', 28, 'Tokyo']
]

# 파일 쓰기
with open('../data/example2.csv', 'w', newline='', encoding='utf-8-sig') as f: # utf-8 BOM: Byte Order Mark (바이트 순서 표식)
    writer = csv.writer(f)
    writer.writerows(csv_data) # csv_data 리스트의 모든 행을 한번에 디스크에 씀.
    # 한줄씩 쓰려면 writer.writerow(row) 를 사용한다.
```

```python
# 파일 읽기 - 한 줄씩 읽기

with open('../data/example2.csv', 'r', encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    # next(reader) 헤더를 건너 뛰려면 이 행을 추가
    for row in reader:
        print(row)
```

```python
# 파일 읽기 - 전체 읽기

with open('../data/example2.csv', 'r', encoding='utf-8-sig') as f:
    content = f.read()
    print(content)
```

```python
# 인코딩 변환

with open('../data/example2-without-bom.csv', 'w', newline='', encoding='utf-8') as f:
    f.write(content)
```

### Regular Expression 정규 표현식

```python
import re

text = '문의사항은 010-1234-5678 또는 support@example.com 으로 연락주세요. 010-5678-1234'

# 전화번호 패턴 추출
mobile_pattern = r'\d{3}-\d{4}-\d{4}' # \d{3,4} 는 숫자 3자리 또는 4자리를 의미
mobile_numbers = re.findall(mobile_pattern, text)
print('추출된 전화번호: ', mobile_numbers)

# 이메일 패턴 추출
email_pattern = r'[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+'
email_address = re.findall(email_pattern, text)
print('추출된 이메일: ', email_address)
```
