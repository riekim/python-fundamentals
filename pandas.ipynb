### 조건으로 데이터 선택
- 데이터프레임객체명.loc[조건문, 컬럼라벨(리스트, 슬라이스 등)]

```python
df.loc[df['국어'] == 100]
```

```python
df.loc[(df['국어'] == 100) & (df['수학'] == 100) & (df['영어'] == 100)]
```

```python
df.loc[(df['국어'] == 100) | (df['수학'] == 100) | (df['영어'] == 100)]
```

```python
# 1. 평균 컬럼 추가 후 소수점 첫째 자리까지 반올림 (.round(1)을 붙여줍니다!)
df['평균'] = df[['국어', '수학', '영어']].mean(axis=1).round(1)

# 2. 스코어(scores)라는 새로운 데이터프레임 변수에 저장
scores = df

# 결과 확인
scores
```
